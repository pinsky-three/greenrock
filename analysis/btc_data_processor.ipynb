```python



# ensure the processed folder exists
os.makedirs("../processed_btc_data", exist_ok=True)

for file in os.listdir("CryptoLM-Bitcoin-BTC-USDT/data/"):
    if file.endswith(".parquet"):
        df = pd.read_parquet(f"../CryptoLM-Bitcoin-BTC-USDT/data/{file}")
        df["month"] = pd.to_datetime(df["timestamp"], unit="ms").dt.strftime("%Y-%m")
        print(f"{file} {df['timestamp'].count()}")

        # save into new processed folder
        df.to_parquet(f"../processed_btc_data/{file}")


# df = pd.read_parquet("../CryptoLM-Bitcoin-BTC-USDT/data/2021-01.parquet")

df.dtypes
```
